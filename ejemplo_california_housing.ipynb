Lo mismo, pero con ggplot2, primero se carga el modulo ggplot, luego se pasa la base de datos de formato ancho a largo (from wide to long)

```r
library(ggplot2)
```

```r
library(tidyr)
library(dplyr)
names(cali)
```

```r
 cali2 = cali %>%
  select('housing_median_age','total_rooms','total_bedrooms','population','households','median_income','median_house_value')
```

```r
View(cali2)
```

```r
cali3 = cali2 %>%
  pivot_longer(cols = c('housing_median_age','total_rooms','total_bedrooms','population','households','median_income','median_house_value'))
  names(cali3)
```

```r
View(cali3)
```

```r
ggplot(cali3, aes(value)) + geom_histogram() + facet_wrap(~name,ncol=3, scales = "free")
```

```r
ggplot(cali3, aes(value,fill = name)) +
  geom_histogram() +
    facet_wrap(~name,ncol=3, scales = "free") +
      theme_bw() + theme(legend.position= "none" )
```

```r
options(repr.plot.width = 12, repr.plot.height = 12)
ggplot(cali3, aes(value,fill = name)) +
  geom_histogram() +
    facet_wrap(~name,ncol=3, scales = "free") +
      theme_bw() + theme(legend.position= "none" )
```
