R tiene una estructura básica que se conoce como el vector, en la celda siguiente, el vector c() se asigna al objeto PesoLechon, y luego se ordena con la función sort

```r
PesoLechon=c(1.5,1.6,1.8,1.2,1.3,1.3,1.3,1.1)
PesoLechon=sort(PesoLechon)
PesoLechon
```

Luego, sobre el objeto PesoLechon (que es un vector) se aplican las funciones mean() y median()

```r
mean(PesoLechon)
median(PesoLechon)
```

Utilizando la función table(), se crea el objeto TablaDeFrecuencias, a continuación, se muestra la tabla de frecuencia del objeto PesoLechon, y luego se escoge del objeto TablaDeFrecuencias la categoria que tenga mayor frecuencia y esta categoria es la moda

```r
TablaDeFrecuencias=table(PesoLechon)
TablaDeFrecuencias
moda=which.max (TablaDeFrecuencias)
moda
```

La función install.packages() instala el modulo "psych" que tiene funciones de estadistica descriptiva como geometric.mean() y harmonic.mean(), que son aplicadas al objeto PesoLecho

```r
install.packages("psych")
library (psych)
geometric.mean(PesoLechon)
harmonic.mean(PesoLechon)
```

La función min() calcula el mínimo, max() calcula el máximo, la función range() calcula el rango, la función var() calcula la varianza, y la función sd() calcula la desviación estándar.

```r
min(PesoLechon)
max(PesoLechon)
range(PesoLechon)
var(PesoLechon)
sd(PesoLechon)
```

El error estándar o típico es el cociente de la desviación estándar, sobre la raíz cuadrada del total de observaciones.


```r
error.standar= sd(PesoLechon)/sqrt(length(PesoLechon))
error.standar
```

El coeficiente de variación es el cociente de la desviación estándar, sobre la media multiplicado por cien


```r
c.v.= (sd(PesoLechon)/mean(PesoLechon))*100
c.v.
```

La función summary() entrega estadística descriptiva para variables continuas

```r
summary(PesoLechon)
```

La funcion quantile() calcula cuantiles

```r
quantile(PesoLechon)
```

Ejemplo para comparar dos vectores que contienen pesos de lechones, se crean los dos vectores

```r
PesoLechon=c(1.5,1.6,1.8,1.2,1.3,1.3,1.3,1.1)
PesoLechon2=c(1.3,1.4,1.5,1.5,1.5,1.8,2)
PesoLechon
PesoLechon2
```

Prueba de t, para muestras independientes, asumiendo varianzas iguales

```r
t.test(PesoLechon, PesoLechon2, var.equal=TRUE)
```

Dos grupos con espesor de grasa de cadera, LoteA y LoteB

```r
LoteA=c(4.3,5.4,6.5,5.5,5.5,6.8,5.2,5.3)
LoteB=c(5.5,5.6,5.8,5.2,5.3,5.3,5.3,5.1,5.1)
LoteA
LoteB
```

Prueba de hipótesis de comparación de varianzas

```r
var.test(LoteA, LoteB)
```

Prueba de t, para comparar promedios con varianzas diferentes

```r
t.test(LoteA, LoteB, var.equal=FALSE)
```

Dos colectas de semen, en dias distintos, de seis toros

```r
Esperma1=c(1.8,1.7,1.6,1.5,1.4,1.2)
Esperma2=c(1.4,1.1,1.1,1.1,1.2,0.8)
BUFALO=c("Iterbio", "Procedonio","Epitacio", "Holmio", "Coltan","Europio")
REPRO=rbind (BUFALO, Esperma1, Esperma2)
REPRO
```

Prueba de t, para comparar promedios de dos poblaciones dependientes

```r
t.test(Esperma1, Esperma2, paired=TRUE)
```
